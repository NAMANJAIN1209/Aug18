```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
path = 'C:\\Users\\namanjain8\\Documents\\Python Scripts_Composite Analysis'
filename = 'Gaussian_Distribution'
df = pd.read_csv(path+'\\'+filename+'.csv')
df.rename(columns={'0':'ND_random'},inplace=True)
z_score=stats.zscore(df['ND_random'])
age = sorted(round(2*z_score + 20,0).astype(int))
devop = sorted(round(200*z_score + 3000,0).astype(int))
javadop = sorted(round(250*z_score + 3000,0).astype(int))
plt.plot(age,devop,label="All devep")
plt.plot(age,javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
plt.show()
```
